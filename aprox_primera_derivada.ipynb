**TAREA 4: APROXIMACIÓN A LA PRIMERA DERIVADA**

Hacer un programa en python para aproximar la primera derivada de la función x(t) = t^2  usando la ecuación de la 1a diferencia finita.

```python
import matplotlib.pyplot as plt
import math
```

```python
#función x(t)
def x(t):
  return math.pow(t,2)
```

```python
#funcion primera derivada
def primer_derivada_aprox(t,h):
  return (x(t+h)-x(t)) /h
```

```python
t_inicial = 0
input_t_fin = input("Introduce el tiempo final para la aproximación de la derivada: ")
t_final = int(input_t_fin)
input_delta = input("Introduce el paso de tiempo(delta): ")
delta = float(input_delta)
```

```python
puntos = [i for i in range(t_inicial, t_final+1)]
```

```python
puntos
```

```python
aproximaciones = [primer_derivada_aprox(t_punto,delta) for t_punto in puntos]
```

```python
valores_x = [x(t_punto) for t_punto in puntos]
```

```python
plt.plot(puntos, valores_x)
plt.xlabel('t')
plt.title(" x(t) = t^2")
plt.grid(True)
plt.show()
```

```python
plt.plot(puntos, aproximaciones)
plt.xlabel('t')
plt.ylabel("Aproximación de la primera derivada")
plt.title("Aproximación de la primera derivada de x(t) = t^2")
plt.grid(True)
plt.show()
```
